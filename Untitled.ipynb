```python
# Example from numpy documentation
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import radon
from skimage.data import shepp_logan_phantom

image = shepp_logan_phantom()

fig, ax = plt.subplots(1, 2)

nx = 500

# Note how there is a symmetry every 180 degrees
theta = np.linspace(0., 360., max(image.shape), endpoint=False)
sinogram = radon(image, theta=theta)
ax[1].imshow(sinogram, extent=(0, 2*np.pi, 0, 360*np.sqrt(2)), aspect='auto')

fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import radon, iradon
from skimage.draw import rectangle

fig, ax = plt.subplots(1, 2)

nx = 512

ii, jj = rectangle((nx//3, nx//2), end=(2*nx//3, 5*nx//6))
image = np.zeros((nx, nx))
image[ii, jj] = 1

theta = np.linspace(0., 180., nx, endpoint=False)

ax[0].imshow(image, extent = (-1, 1, -1, 1))

f = radon(image, theta=theta)

ax[1].imshow(f, extent = (0, 2*np.pi, 0, 1), aspect='auto')
```
